```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
dataset = pd.read_csv('C:/Users/Ahmadi/OneDrive/Desktop/datasets/2.1 movies.csv')
dataset
```

```python
items = dataset['genres']
items = items.str.split('|')
print(items)
```

```python
te = TransactionEncoder()
#transform the input dataset into a one-hot encoded Numpy boolean array
te_ary = te.fit(items).transform(items)
df = pd.DataFrame(te_ary,columns=te.columns_)
#using apriori algorithm with minimum support value=0.1
frequent_items = apriori(df,min_support=0.1,use_colnames=True)
#Generating frequent values
frequent_items
```

```python
from mlxtend.frequent_patterns import association_rules
#generating association rules with confidence > 0.2
association_rules(frequent_items,metric='confidence',min_threshold=0.2)
```

```python
#generating association rules with lift > 0.2
association_rules(frequent_items,metric='lift',min_threshold=0.2)
```
