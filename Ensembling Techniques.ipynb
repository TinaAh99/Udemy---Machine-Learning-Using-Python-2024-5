```python
import pandas as pd
url = 'C:/Users/Ahmadi/OneDrive/Desktop/datasets/2.1 Earnings_Manipulation220.csv'
df = pd.read_csv(url)
df.columns = df.columns.map(str.lower)
df
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, cohen_kappa_score
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
def classify_report(ytest,y_pred):
    cm = confusion_matrix(ytest,y_pred)
    print("Confusion Matrix:\n",cm)
    error_rate = 1 - accuracy_score(ytest,y_pred)
    print("classification report:\n")
    print(classification_report(ytest,y_pred))
    print('Accuracy:\n',accuracy_score(ytest,y_pred))
    print('ErrorRate:\n',error_rate)
    print('Cohen Kappa score:\n',cohen_kappa_score(ytest,y_pred))
```

```python
df['year ending'] = pd.to_datetime(df['year ending'])

# استخراج سال، ماه و روز و اضافه کردن آن‌ها به ابتدای دیتافریم
df.insert(1, 'year_ending', df['year ending'].dt.year)
df.insert(2, 'month_ending', df['year ending'].dt.month)
df.insert(3, 'day_ending', df['year ending'].dt.day)
df.drop(columns=['year ending'], inplace=True)
df
```

```python
# prediction with supervised algorithm

#input features
x = df.iloc[:,:12].values
#output class
y = df.iloc[:,12].values

#splitting the data into training and test
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=0)

#1)naive bayes algorithm
#create a Gaussian naive bayes model
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
#fit a model
gnb.fit(xtrain,ytrain)
y_pred_gnb = gnb.predict(xtest)
classify_report(ytest,y_pred_gnb)
```

```python
#2)KNN
from sklearn.neighbors import KNeighborsClassifier
# KNN model creation 
knn = KNeighborsClassifier(n_neighbors=5)

#fit model
knn.fit(xtrain,ytrain)
#perform predictions
y_pred_knn = knn.predict(xtest)

classify_report(ytest,y_pred_knn)
```

```python
#ensembling technique AdaBoost
from sklearn.ensemble import AdaBoostClassifier
seed=7
num_trees = 100
model = AdaBoostClassifier(n_estimators=num_trees,random_state=seed)
model.fit(xtrain,ytrain)
y_pred_adb = model.predict(xtest)
classify_report(ytest,y_pred_adb)
```

```python
# ensembling technique Gradient Boosting or XG Boost
from sklearn.ensemble import GradientBoostingClassifier
seed = 7
num_trees = 500
model = GradientBoostingClassifier(n_estimators=num_trees,random_state=seed,max_depth=20)
model.fit(xtrain,ytrain)
y_pred_adb = model.predict(xtest)
classify_report(ytest,y_pred_adb)
```
