```python
import pandas as pd

#csv file
url = 'C:/Users/Ahmadi/OneDrive/Desktop/datasets/1.1 titanic.csv'

#Creating a dataframe
df = pd.read_csv(url).fillna(0)

print('Max:',df['Age'].max())
print('Min:',df['Age'].min())
print('sum:',df['Age'].sum())
print('mean:',df['Age'].mean())
print('median:',df['Age'].median())
print('mode:',df['Age'].mode())
print('count:',df['Age'].count())
print('var:',df['Age'].var())
print('std:',df['Age'].std())
```

```python
#concating two dataframe
df1 = df.iloc[0:5]
print(df1)
df2 = df.iloc[5:10]
print(df2)
df3 = pd.concat([df1,df2],axis=0)
print(df3)
df4 = pd.concat([df1,df2],axis=1)
print(df4)
```

```python
#mergind dataframes
employee_data = {'employee_id':['1','2','3','4'],'name':['tina ahmadi','theo james','nicole kidman','majid asadi']}
df_employee = pd.DataFrame(employee_data,columns=['employee_id','name'])
sales_data = {'employee_id':['1','2','3','4'],'total_sale':[23456,2512,2345,1455]}
df_sales = pd.DataFrame(sales_data,columns=['employee_id','total_sale'])
print(pd.merge(df_employee,df_sales,on='employee_id'))
```

```python
print(pd.merge(df_employee,df_sales,on='employee_id',how='outer'))
```
